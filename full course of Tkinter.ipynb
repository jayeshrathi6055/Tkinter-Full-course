```python
from tkinter import *
from PIL import Image,ImageTk
root = Tk()
# change width * height of screen------------------->
root.geometry("644x434")
root.minsize(200,100)
root.maxsize(13600,800)
# Making a function
def getvals():
    print("it works")
    print(f"{name_value.get(), phone_value.get(), gender_value.get(),emergency_value.get(), paymentmode_value.get(), foodservice_value.get()} ")
    with open("Travel.txt","a") as f:
        f.write(f"{name_value.get(), phone_value.get(), gender_value.get(), emergency_value.get(), paymentmode_value.get(), foodservice_value.get()}\n")

# Heading
Label(root,text = "WELCOME TO TRAVELS", font = "conicsanas 13 bold", pady = 15).grid(row = 0,column = 3)
# Text values
name = Label(root,text = "Name")
phone = Label(root,text = "Phone")
gender = Label(root,text = "Gender")
emergency = Label(root,text = "Emergency")
paymentmode = Label(root,text = "Payment mode")
# pack values
name.grid(row = 1,column =2 )
phone.grid(row = 2,column =2 )
gender.grid(row = 3,column = 2)
emergency.grid(row = 4,column = 2)
paymentmode.grid(row = 5,column = 2)
# make variable
name_value = StringVar()
phone_value = IntVar()
gender_value = StringVar()
emergency_value = IntVar()
paymentmode_value = StringVar()
foodservice_value = IntVar()
# entries for a phone
name_entry = Entry(root,textvariable = name_value )
phone_entry = Entry(root,textvariable = phone_value)
gender_entry = Entry(root,textvariable = gender_value)
emergency_entry = Entry(root,textvariable = emergency_value)
paymentmode_entry = Entry(root,textvariable = paymentmode_value)
# packing all entries
name_entry.grid(row =1 ,column = 3)
phone_entry.grid(row = 2,column = 3)
gender_entry.grid(row = 3,column = 3)
emergency_entry.grid(row =4 ,column = 3)
paymentmode_entry.grid(row =5 ,column = 3)
# Checkbox
foodservice = Checkbutton(text = "Want to prebook your meal",variable = foodservice_value)
foodservice.grid(row = 6, column = 3)
# Button and packing it and assign it a command
Button(text = "Submit",command = getvals).grid(row = 7, column = 3)

root.mainloop()
```

```python
from tkinter import *
from PIL import Image, ImageTk
import tkinter.messagebox as tmsg

root = Tk()
# change width * height of screen------------------->
root.geometry("644x434")
root.minsize(200, 100)
root.maxsize(13600, 800)


def func():
    print("hello")

def help():
    print("I will help you")
    tmsg.showinfo("Help","Help with gui")

def rate():
    print("Rate us")
    tmsg.askquestion("Rating","Is your experiance is good")

def extract():
    print("Extract file")
    tmsg.askretrycancel("Extract file","Your file is not opening.")
    
# Make message box
menubar = Menu(root)
m1 = Menu(menubar, tearoff=0)  # tearoff can be 0 or 1
m1.add_command(label="New File")
m1.add_command(label="open", command=func)
m1.add_command(label="save", command=func)
m1.add_command(label="save as", command=func)
m1.add_separator()  # it is used to separate the options.
m1.add_command(label="Help", command=help)
m1.add_command(label="rate", command=rate)
m1.add_command(label="Extract", command=extract)
root.config(menu=menubar)
menubar.add_cascade(label="File", menu=m1)

root.mainloop()
```

```python
from tkinter import *
root = Tk()
# change width * height of screen------------------->
root.geometry("644x434")
root.minsize(200, 100)
root.maxsize(13600, 800)

# Make a List Box
i=0
def add():
    global i
    lbx.insert(ACTIVE,f"{i}")     # ACTIVE means enter value above of selected line
    i+=1

lbx = Listbox(root)
lbx.pack(pady = 15)
lbx.insert(END,"First")
Button(root,text = "Add item",command = add).pack()

root.mainloop()
```

```python
from tkinter import *
root = Tk()
root.geometry("500x400")

# Use of update------------------------>
def update():
    string.set("Buffering.....")
    a.update()                     # Through this function can update the value
    import time
    time.sleep(3)
    string.set("Start")

string = StringVar()
string.set("start")

a = Label(root,textvariable = string,bg = "black",fg = "white",font = "comicsansns 15 bold")
a.pack(fill = X)
Button(root,text = "GUI",command = update).pack(pady = 20)

root.mainloop()
```

```python
from tkinter import *

# Store GUI in class
class GUI(Tk):
    def __init__(self):
        super().__init__()
        self.geometry("600x400")

    def click(self):
        print("done")

    def button(self):
        Button(text = "click me",command = self.click).pack()

window = GUI()
window.button()
window.mainloop()
```
